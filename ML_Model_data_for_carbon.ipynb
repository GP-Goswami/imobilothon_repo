```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.Source = le.fit_transform(df.Source)
df['Steel type'] = le.fit_transform(df[['Steel type']])
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
X=df.drop('Final hardness (HRC) - post tempering',axis=1)
Y=df[['Final hardness (HRC) - post tempering']]
X.shape
# Y.shape
```

```python

X = X.replace('?', 0).astype(float)
Y = Y.replace('?', 0).astype(float)
# X.dropna(inplace=True)
# Y.dropna(inplace=True)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
model = MultiOutputRegressor(RandomForestRegressor(
    n_estimators=200,         # Increase trees for better generalization
    max_depth=15,             # Depth of each tree to avoid overfitting
    min_samples_split=5,      # Minimum samples required to split a node
    min_samples_leaf=2,       # Minimum samples required at each leaf node
    max_features='sqrt',      # Features to consider when looking for the best split
    random_state=42,          # Seed for reproducibility
    n_jobs=-1                 # Parallelize computations for faster training
))
```

```python
model.fit(X_train, y_train)
```

```python
testmodel=model.predict(X_test)
# print(testmodel)
output=model.predict([['0','6',	'0',	'600',	'204.4',	'0.25',	'0.79',	'0.012',	'0.026',	'0.11',	'0.0',	'0.00',	'0.0',	'0.0',	'0.0',	'0.0']])
print(output)
```

```python
r2 = r2_score(y_test, testmodel)
print(f"R-squared: {r2}")

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, testmodel)
print(f"Mean Squared Error: {mse}")

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
```
